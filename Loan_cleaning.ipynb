```python
import pandas as pd
```

```python
df= pd.read_csv("D:/loan.csv",low_memory=False)
```

```python
df.head()
```

```python
df['int_rate'] = df['int_rate'].str.replace('%','').astype(float)
```

```python
df['term'] = df['term'].str.extract('(\d+)').astype(int)
```

```python
important_cols = [
    # Identifiers
    'id','member_id',
    
    # Core exposure
    'loan_amnt',
    'funded_amnt',

    # Pricing
    'int_rate',
    'term',

    # Risk segmentation
    'grade',
    'sub_grade',

    # Borrower strength
    'annual_inc',
    'dti',

    # Behavioral credit risk
    'delinq_2yrs',
    'inq_last_6mths',
    'pub_rec_bankruptcies',

    # Product segmentation
    'purpose',

    # Geography (for exposure concentration)
    'addr_state',

    # Time (for vintage analysis)
    'issue_d',

    # Performance
    'loan_status',

    # Financial outcome
    'total_rec_prncp',
    'recoveries'
]

```

```python
df= df[important_cols]
```

```python
df
```

```python
df.info()
```

```python
# Keep only closed loans
df = df[df['loan_status'].isin(['Fully Paid','Charged Off'])].copy()

# Create default flag
df['default_flag'] = df['loan_status'].map({
    'Fully Paid': 0,
    'Charged Off': 1
})
```

```python
df['default_flag'].value_counts(normalize=True)
```

```python
df.head()
```

```python
df.columns
```

```python
df['pub_rec_bankruptcies'] = df['pub_rec_bankruptcies'].fillna(0)
```

```python
df['issue_d'] = df['issue_d'].str.strip()
df['issue_d'] = pd.to_datetime(df['issue_d'], format='%b-%y', errors='coerce')
```

```python
int_cols = [
    'delinq_2yrs',
    'inq_last_6mths',
    'pub_rec_bankruptcies',
    'default_flag'
]

for col in int_cols:
    df[col] = df[col].fillna(0).astype(int)
```

```python
# Final sanity check
print("Final shape:", df.shape)
print(df.isnull().sum())

# Export to CSV
df.to_csv("loan_cleaned.csv", index=False)
```
